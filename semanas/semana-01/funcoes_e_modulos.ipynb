## *Parâmetros*

```python
#Sintaxe de função com parâmetros
def funcao_com_parametros(x, y): # nesse caso, as variáveis x e y são os parâmetros da função
  print(f'{x} e {y} são os parâmetros dessa função.')
```

```python
#Chamada de função com parâmetros
funcao_com_parametros(6, 7)
```

```python
# é sempre necessário fazer a chamada da função com
# a quantidade correta de parâmetros da mesma.
```

```python
#É possível utilizar o *args caso a quantidade de parâmetros da função
#não esteja bem definida.
#Nesse caso, o parâmetro passa a ser uma tupla, que dependerá da chamada.
def funcao_n_parametros(*parametros):
  print(parametros)
  print(type(parametros))
```

```python
funcao_n_parametros(10)
```

```python
funcao_n_parametros('a', 'b', 'c')
```

```python
funcao_n_parametros(True, False)
```

```python
#É também possível chamar funções que não contenham
#uma quantidade definida de argumentos nomeados, com os chamados **kwargs
def funcao_kwargs(**argumentos):
  print(argumentos)
  print(type(argumentos))
```

```python
funcao_kwargs(nome = 'Arthur', idade = 21, altura = 1.95)
```

```python
#Como o parâmetro assume o tipo de um dicionário, é possível
#realizar todas as operações da estrutura.
def funcao_kwargs(**argumentos):
  for x in argumentos.keys():
    print(x)
```

```python
funcao_kwargs(nome = 'Arthur', idade = 21, altura = 1.95)
```

```python
#Utilizando um valor padrão caso o usuário não envie o parâmetro
def imprime_time(time = 'Nenhum time recebido.' ):
  print(time)
```

```python
imprime_time()
```

# **LIST COMPREHENSION**

Utilizado para criar listas a partir de iteradores em uma única linha de código.

Substitui a necessidade de fazer loops e adicionar código.

```python
#Sintaxe de list comprehension
[expressão for item in iterável]
```

```python
#Criando uma lista com o dobro de 10 a 15

#Jeito tradicional
lista_quadrados = []
for i in range(10, 16):
  lista_quadrados.append(i * 2)
print(lista_quadrados)
```

```python
#Utilizando list comprehension
#Retomando a sintaxe: [expressão for item in iterável]
quadrados = [i * 2 for i in range(10, 16)]

print(quadrados)
```

```python
#Utilizando if e else com list comprehension

#Criando uma lista com números múltiplos de 4 entre 4 e 50
#Jeito tradicional
multiplos_quatro = []
for i in range(4, 51):
  if i % 4 == 0:
    multiplos_quatro.append(i)


print(multiplos_quatro)
```

```python
#Utilizando list comprehension
multiplos = [i for i in range(4, 51) if i % 4 == 0]

print(multiplos)
```
