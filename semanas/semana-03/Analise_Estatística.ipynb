# **Análise estatística sobre a saúde do sono**

O dataset escolhido apresenta fatores que podem estar relacionados a qualidade do sono.

O dataset possui diversas colunas que se relacionam com o tema, mas farei as métricas de acordo com 2 delas, sendo essas:


*   **Quality of Sleep:** qualidade do sono
*   **Stress Level:** nível de estresse

O dataset foi adicionado previamente ao Colab e pode ser acessado pelo link: https://www.kaggle.com/datasets/uom190346a/sleep-health-and-lifestyle-dataset



```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
df = pd.read_csv('sleep_health.csv') #coletando o dataset
```

```python
print('Amostra do dataset:')
df.head()
```

## *Métricas Descritivas*

### Média

```python
print("* Média aritmética das variáveis *")

media_qualidade_sono = df['Quality of Sleep'].mean()
print(f"Qualidade do sono: {media_qualidade_sono:.2f}")

media_estresse = df['Stress Level'].mean()
print(f"Nível de estresse: {media_estresse:.2f}")
```

### Mediana

```python
print("* Mediana das variáveis *")

mediana_qualidade_sono = df['Quality of Sleep'].median()
print(f"Qualidade do sono: {mediana_qualidade_sono:.2f}")

mediana_estresse = df['Stress Level'].median()
print(f"Nível de estresse: {mediana_estresse:.2f}")
```

### Moda

```python
print("* Moda das variáveis *")

moda_qualidade_sono = df['Quality of Sleep'].mode()[0]
print(f"Qualidade do sono: {moda_qualidade_sono}")

moda_estresse = df['Stress Level'].mode()[0]
print(f"Nível de estresse: {moda_estresse}")
```

## *Médidas de Dispersão*

### Amplitude

Diferença entre o maior e menor valor presentes na coluna.

```python
print("* Amplitude das variáveis *")

ampl_qualidade_sono = df['Quality of Sleep'].max() - df['Quality of Sleep'].min()
print(f"Qualidade do sono: {ampl_qualidade_sono}")

ampl_estresse = df['Stress Level'].max() - df['Stress Level'].min()
print(f"Nível de estresse: {ampl_estresse}")
```

### Variância
Medida que mostra o quão distante os valores estão da média.

```python
print("* Variância dos valores *")

var_qualidade_sono = df['Quality of Sleep'].var()
print(f"Qualidade do sono: {var_qualidade_sono:.2f}")

var_estresse = df['Stress Level'].var()
print(f"Nível de estresse: {var_estresse:.2f}")
```

### Desvio Padrão
Mostra em média o quanto os valores se distanciam da média.

O desvio padrão é calculado como a raiz quadrada positiva da variância.

```python
print("* Desvio Padrão dos valores *")

std_qualidade_sono = df['Quality of Sleep'].std()
print(f"Qualidade do sono: {std_qualidade_sono:.2f}")

std_estresse = df['Stress Level'].std()
print(f"Nível de estresse: {std_estresse:.2f}")
```

## *Teste de Hipótese*

**Pergunta:** A qualidade média do sono é significativamente menor em indivíduos com altos níveis de estresse?

**Hipótese Nula (H0):** A qualidade do sono não se altera, o estresse não tem efeito.

**Hipótese Alternativa (H1):** A qualidade do sono se altera entre os indivíduos com altos níveis de estresse e os de baixo nível.

```python
correlacao, valor_p = stats.pearsonr(df['Stress Level'], df['Quality of Sleep'])

print(f"Coeficiente de Correlação (r): {correlacao:.4f}")
print(f"p-valor: {valor_p:.4f}")
```

Rejeita-se a hipótese nula e se comprova uma relação inversamente proporcional entre o nível de estresse e a qualidade do sono.
