## *Operações com arrays*

```python
#Operações Aritméticas - as operações são realizadas elementos por elemento entre os arrays
array1 = np.array([1,2,3,4])
array2 = np.array([4,3,2,1])

print(array1)
print(array2)

#Soma
soma = array1 + array2
print(f'Soma de {array1} com {array2}: {soma}')

#Multiplicação
mult = array1 * array2
print(print(f'Produto de {array1} com {array2}: {mult}')
)

#Divisão
div = array1 / array2
print(f'Razão de {array1} com {array2}: {div}')
```

```python
#Operação com Escalar - opera todos os elementos do array com o escalar
#É possível utilizar qualquer operação aritmética
array = np.array([58,62,70,94,2002])

escalar_array = array * 6
print(escalar_array)
```

```python
#Broadcasting - operações entre arrays de diferentes dimensões
array1 = np.array([10,20,30])
array2 = np.array([[1,3,5],[2,4,6]])

print(f'Array1: {array1}')
print(f'Array2: {array2}')

soma = array1 + array2
print(soma)
```

```python
#No broadcasting, o numpy adequa a estrutura com menos dimensões, espalhando seus valores virtualmente
```

## *Funções de agregação*
Funções que resumem os dados presentes no array

```python
#Função que soma todos os elementos do array - sum()
array1 = np.array([1,2,3,4,5])
soma1 = array1.sum()
print(f'Soma dos elementos de {array1}: {soma1}')
```

```python
#Soma por colunas ou linhas de uma matriz - .sum(axis=0 ou 1) -> 0 para colunas e 1 para linhas
matriz = np.array([[5,10,15,20],[25,30,35,40]])
print(matriz)
soma_coluna = matriz.sum(axis=0)
print(f'Soma das colunas: {soma_coluna}')

soma_linhas = matriz.sum(axis=1)
print(f'Soma das linhas: {soma_linhas}')
```

```python
#Função que calcula a média aritmética dos elementos do array - mean()
array2 = np.array([10,20,30,40])
media = array2.mean()
print(f'A média da matriz {array2} é: {media}')
```

```python
#Da mesma forma que o sum, é possível pegar a média somente das
#colunas ou das linhas da matriz, utilizando o axis = 0 ou = 1
media_colunas = matriz.mean(axis = 0)
print(f'Média das colunas da matriz {matriz}: {media_colunas}')
media_linhas = matriz.mean(axis = 1)
print(f'Média das linhas da matriz {matriz}: {media_linhas}')
```

```python
#Função que calcula o desvio padrão dos elementos - std.()
array3 = np.array([15,30,45,60])
desvio = array3.std()
print(desvio)
```

```python
#Funções para encontrar o valor mínimo e máximo do array - min() e max()
min = array3.min()
print(min)
max = array3.max()
print(max)
```

```python
#Função que retorna o ÍNDICE do maior elemento presente no array
array4 = np.array([[10,20,30,40],[50,80,60,70]])
ind_maior = array4.argmax()
print(ind_maior)
```

## *Indexação Booleana*

```python
#Combinando condições
dados = np.arange(1,11)
print(dados)
condicao1 = dados % 2 == 0
condicao2 = dados < 5

dados_filtrados = dados[condicao1 & condicao2] #somente elementos pares e menores que 5
print(dados_filtrados)
```

## *Visão x Cópia*
O fatiamento de um array gera a chamada view (visão) do mesmo, o que significa que cada alteração no array fatiado também será executada no array original.
Para que não ocorra alterações indesejadas no original, é possível criar cópias do mesmo.

```python
#Exemplificando a visão de um array
array = np.array([1,2,3,4,5])
print(array)

visao = array[2:4] #fatiando o array
print(visao)

visao[0] = 100 #alterando o primeiro elemento do array fatiado
print(visao)
print(array) #o array original também sofre a alteração
```

```python
#Criando uma cópia com o método .copy()
array_copia = array.copy()
print(array_copia)

array_copia[0] = 90 #ao modificar o array copiado, isso não afeta o original
print(array_copia)
print(array)
```

## *Manipulação de Shapes*
Processo de alteração das dimensões de um array sem que haja perda de dados.

### Função .reshape()

Permite a alteração da forma de um array, desde que o novo array possua a mesma quantidade de elementos que o original.

```python
array = np.arange(1, 11) #array unidimensional com 10 elementos
print(array)

#Transformando em uma matriz
#É necessário que a matriz possua a mesma quantidade de elementos
#Nesse caso vou usar uma matriz de 2 linhas e 5 colunas
matriz = array.reshape(2, 5)
print(matriz)
```

```python
#O parâmetro -1 pode ser utilizado para o cálculo automático
#do eixo restante, desde que seja possível remanejar a estrutura
#com o parâmetro enviado
matriz2 = array.reshape(5, -1)
print(matriz2)
```

```python
#Código que gera erro, pois não é possível fazer uma matriz
#de 10 elementos com 3 linhas
matriz2 = array.reshape(3, -1)
print(matriz2)
```

```python
#Em suma, o produto dos n parâmetros deve ser igual a quantidade
#de elementos presentes na estrutura original
```

### Função newaxis

Adiciona uma nova dimensão ao array.

```python
array = np.arange(1, 6)
print(f'Array original: {array}')
print(f'Shape do array original: {array.shape}')

#Adicionando uma coluna ao array
array_coluna = array[:,np.newaxis] #nessa linha, primeiro eu pego todos os elementos do array, e posteriormente crio a nova coluna
print(f'Array coluna: {array_coluna}')
print(f'Shape do array coluna: {array_coluna.shape}')

#Adicionando uma linha ao array
array_linhas = array[np.newaxis,:] #resulta em um array com 1 linha e 5 colunas
print(f'Array com 2 dimensões: {array_linhas}')
print(f'Shape do array: {array_linhas.shape}')
```

### Função .expand_dims()

Função que adiciona uma nova dimensão em um eixo específico.

```python
#Sintaxe - np.expand_dims(array, axis = eixo)
#A nova dimensão é especificada por meio do parâmetro axis = eixo
array = np.arange(1, 6)
print(array)

array_expandido = np.expand_dims(array, axis = 1) #adicionando uma coluna ao array
print(array_expandido)
print(array_expandido.shape)
```

### Função .squeeze()


Remove dimensões do array.

```python
print(array_expandido)
arrayx = np.squeeze(array_expandido)
print(arrayx)
print(arrayx.shape)
```

### Transposição de arrays - Função .T

Inverte as dimensões do array, por exemplo:
Uma matriz M[i,j], vira uma matriz M[j,i]

```python
matriz = np.array([[1,2,3,4], [5,6,7,8]])
print(matriz)
print(matriz.shape)
print()

transposta = matriz.T
print(transposta)
print(transposta.shape)
```

### Achatamento de arrays - Funções .flatten() e .ravel()

Transformam um array de n dimensões em um array unidimensional.

**.flatten():** retorna uma cópia do array original (gasta mais memória)

**.ravel():** retorna uma visão do array original

```python
matriz = np.arange(1,10).reshape(3, 3)
print(matriz)
print(matriz.shape)
print()

#achatando a matriz
achatada = matriz.ravel()
print(achatada)
print(achatada.shape)
```

### Funções .vstack() e .hstack() - Empilhamentos verticais e horizontais

**.vstack()**: caso sejam mútliplos arrays, eles devem possuir o mesmo número de colunas

**.hstack()**: devem possuir o mesmo número de linhas

```python
#Utilizando .vstack
array1 = np.arange(1,6)
array_v = np.vstack(array1)
print(array_v)

array2 = np.arange(6,11)
array_v2 = np.vstack((array1,array2))
print(array_v2)
```

```python
#Utilizando o .hstack()
matriz = np.arange(1,10).reshape(3,3)
print(matriz)
print()
matriz_h = np.hstack(matriz)
print(matriz_h)
```
